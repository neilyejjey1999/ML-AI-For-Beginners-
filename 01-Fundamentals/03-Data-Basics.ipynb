## Let's Load and Explore Data! 💻

```python
# Import libraries we'll use
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create example data: Student information
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [18, 19, 18, 19, 18],
    'Math_Score': [85, 92, 78, 95, 88],
    'Science_Score': [90, 88, 75, 92, 91],
    'Hours_Studied': [5, 8, 3, 9, 6]
}

# Create a DataFrame (table)
df = pd.DataFrame(data)

# Display the first few rows
print("Our dataset:")
print(df)
```

```python
# Get basic info about the dataset
print("Dataset shape (rows, columns):", df.shape)
print("\nColumn names:", df.columns.tolist())
print("\nData types:")
print(df.dtypes)
```

```python
# Get statistical summary
print("Statistical summary:")
print(df.describe())
```

## Accessing Data 🎯

```python
# Access a single column
print("Math scores:")
print(df['Math_Score'])
print("\nMean math score:", df['Math_Score'].mean())
```

```python
# Access a single row
print("Alice's data:")
print(df.iloc[0])  # First row (index 0)
```

```python
# Filter data (find students with Math score > 85)
good_math = df[df['Math_Score'] > 85]
print("Students with Math score > 85:")
print(good_math)
```

## Handling Missing Data 🚨

```python
# Create data with missing values
data_with_missing = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [18, np.nan, 18, 19, 18],
    'Math_Score': [85, 92, np.nan, 95, 88],
    'Science_Score': [90, 88, 75, np.nan, 91]
}

df_missing = pd.DataFrame(data_with_missing)
print("Data with missing values (NaN):")
print(df_missing)
print("\nMissing value count per column:")
print(df_missing.isnull().sum())
```

```python
# Option 1: Remove rows with missing values
df_clean_drop = df_missing.dropna()
print("After removing rows with missing values:")
print(df_clean_drop)
print(f"\nRows remaining: {len(df_clean_drop)} (was {len(df_missing)})")
```

```python
# Option 2: Fill missing values with mean
df_clean_fill = df_missing.copy()
df_clean_fill['Age'].fillna(df_clean_fill['Age'].mean(), inplace=True)
df_clean_fill['Math_Score'].fillna(df_clean_fill['Math_Score'].mean(), inplace=True)
df_clean_fill['Science_Score'].fillna(df_clean_fill['Science_Score'].mean(), inplace=True)

print("After filling with mean values:")
print(df_clean_fill)
print("\nNo more missing values!")
```

## Visualizing Data 📈

```python
# Create a histogram of math scores
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.hist(df['Math_Score'], bins=10, color='blue', edgecolor='black')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.title('Distribution of Math Scores')
plt.grid(True, alpha=0.3)

# Create a scatter plot
plt.subplot(1, 2, 2)
plt.scatter(df['Hours_Studied'], df['Math_Score'], color='red', s=100)
plt.xlabel('Hours Studied')
plt.ylabel('Math Score')
plt.title('Hours Studied vs Math Score')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("Notice: More study hours tend to correlate with higher scores!")
```

## Data Normalization 📏

Sometimes features have very different scales. We should normalize them so the model treats them fairly.

```python
# Example: Age (18-19) vs Math Score (75-95) have very different scales
print("Original data ranges:")
print(f"Age: {df['Age'].min()} - {df['Age'].max()}")
print(f"Math Score: {df['Math_Score'].min()} - {df['Math_Score'].max()}")
print(f"Hours Studied: {df['Hours_Studied'].min()} - {df['Hours_Studied'].max()}")
```

```python
# Normalize to 0-1 range
df_normalized = df.copy()

# For each numeric column, normalize it
numeric_columns = ['Age', 'Math_Score', 'Science_Score', 'Hours_Studied']

for col in numeric_columns:
    min_val = df[col].min()
    max_val = df[col].max()
    df_normalized[col] = (df[col] - min_val) / (max_val - min_val)

print("\nNormalized data (0-1 scale):")
print(df_normalized)
print("\nAll values now between 0 and 1!")
```

## Data Preparation Checklist ✅

Before feeding data to a machine learning model:

1. ✅ **Load the data** - Read from file or database
2. ✅ **Explore it** - Look at shape, types, statistics
3. ✅ **Handle missing values** - Drop or fill NaN values
4. ✅ **Remove outliers** - Sometimes (extreme values can skew learning)
5. ✅ **Normalize/scale** - Put features on similar scales
6. ✅ **Split into train/test** - Keep some data hidden for evaluation
7. ✅ **Ready to train!** - Feed to ML algorithm

## Key Takeaways 🎯

✅ **Pandas DataFrames** are powerful tools for working with tabular data

✅ **Always explore your data** before training a model

✅ **Missing data** must be handled (drop or fill)

✅ **Visualization** helps understand patterns

✅ **Normalization** puts features on fair scales

✅ **Data preparation** takes as much time as model building!
