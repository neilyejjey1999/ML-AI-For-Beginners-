# Linear Regression 📈

Predicting continuous values using straight lines (and hyperplanes). The most foundational ML algorithm.

## What You'll Learn
- How linear regression works
- Simple (1 feature) and multiple (many features) regression
- Fitting and evaluation
- Understanding predictions
- Real-world applications

## The Basic Idea 🎯

Find the best-fit line through data points.

### Simple Linear Regression
**Formula:** y = mx + b

- **y:** Prediction (target)
- **x:** Input feature
- **m:** Slope (how steep)
- **b:** Intercept (where it crosses y-axis)

**Real example:** Predicting house price based on size
- Input (x): Square feet
- Output (y): Price
- Slope: $100 per square foot
- Intercept: $50,000 base price

## Simple Linear Regression 💻

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Create simple dataset: Study hours vs Test Score
hours = np.array([2, 3, 4, 5, 6, 7, 8, 9, 10]).reshape(-1, 1)
scores = np.array([55, 60, 65, 70, 75, 80, 85, 90, 95])

print("Dataset: Study Hours vs Test Scores")
print(f"Hours:  {hours.flatten()}")
print(f"Scores: {scores}")
```

```python
# Train linear regression model
model = LinearRegression()
model.fit(hours, scores)

# Get slope and intercept
slope = model.coef_[0]
intercept = model.intercept_

print(f"\nModel found:")
print(f"  Slope (m):     {slope:.2f}")
print(f"  Intercept (b): {intercept:.2f}")
print(f"\nFormula: Score = {slope:.2f} × Hours + {intercept:.2f}")
print(f"\nInterpretation: Each hour of study increases score by {slope:.1f} points")
```

```python
# Make predictions
predictions = model.predict(hours)

print("Predictions vs Actual:")
print(f"\nHours | Actual | Predicted")
print("-" * 30)
for h, actual, pred in zip(hours.flatten(), scores, predictions):
    print(f"{h:4.0f}  |  {actual:3.0f}  |  {pred:7.2f}")
```

```python
# Visualize
plt.figure(figsize=(10, 6))
plt.scatter(hours, scores, color='blue', s=100, label='Actual data', alpha=0.7)
plt.plot(hours, predictions, color='red', linewidth=2, label='Fitted line')

# Add prediction for 11 hours
new_hours = np.array([[11]])
new_prediction = model.predict(new_hours)[0]
plt.scatter(11, new_prediction, color='green', s=150, marker='*', 
           label=f'Prediction: 11 hours → {new_prediction:.1f} score', zorder=5)

plt.xlabel('Study Hours', fontsize=12)
plt.ylabel('Test Score', fontsize=12)
plt.title('Simple Linear Regression: Study Hours vs Test Score', fontsize=14)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"\nPrediction for 11 hours: {new_prediction:.1f}")
```

## Measuring Fit Quality 📊

```python
# Calculate residuals (errors)
residuals = scores - predictions

print("Residuals (Prediction Error):")
print(f"\nHours | Actual | Predicted | Residual")
print("-" * 45)
for h, actual, pred, res in zip(hours.flatten(), scores, predictions, residuals):
    print(f"{h:4.0f}  |  {actual:3.0f}  |  {pred:9.2f}  |  {res:+7.2f}")

print(f"\nSum of residuals: {sum(residuals):.4f} (should be ~0)")
```

```python
# Metrics
mse = mean_squared_error(scores, predictions)
rmse = np.sqrt(mse)
mae = np.mean(np.abs(residuals))  # Mean Absolute Error
r2 = r2_score(scores, predictions)

print("\n📊 Evaluation Metrics:")
print(f"\nMean Squared Error (MSE): {mse:.2f}")
print(f"  → Average squared error")
print(f"  → Lower is better")

print(f"\nRoot Mean Squared Error (RMSE): {rmse:.2f}")
print(f"  → Square root of MSE")
print(f"  → In same units as target (scores)")
print(f"  → Interpretation: Average prediction error ≈ ±{rmse:.1f} points")

print(f"\nMean Absolute Error (MAE): {mae:.2f}")
print(f"  → Average absolute error")
print(f"  → Also in units of target")

print(f"\nR² Score: {r2:.4f}")
print(f"  → Proportion of variance explained")
print(f"  → Range: 0 to 1 (higher is better)")
print(f"  → 1.0 = Perfect fit")
print(f"  → Our model explains {r2*100:.1f}% of variation")
```

## Multiple Linear Regression 🏠

```python
# House price dataset with multiple features
houses = {
    'sqft': [1000, 1200, 1500, 1800, 2000, 2200, 2500, 2800, 3000, 3200],
    'bedrooms': [2, 2, 3, 3, 4, 3, 4, 4, 5, 5],
    'age': [20, 5, 15, 2, 10, 1, 8, 3, 12, 0],
    'price': [150000, 200000, 210000, 280000, 300000, 320000, 380000, 420000, 450000, 500000]
}

df = pd.DataFrame(houses)
print("House Features:")
print(df)

X = df[['sqft', 'bedrooms', 'age']]
y = df['price']
```

```python
# Train multiple regression
model_multi = LinearRegression()
model_multi.fit(X, y)

# Coefficients
print("\nMultiple Regression Model:")
print(f"Price = {model_multi.intercept_:,.0f}", end="")
for i, col in enumerate(X.columns):
    coef = model_multi.coef_[i]
    sign = '+' if coef >= 0 else '-'
    print(f" {sign} {abs(coef):,.0f} × {col}", end="")
print()

print("\n\nCoefficient Interpretation:")
for col, coef in zip(X.columns, model_multi.coef_):
    print(f"  {col:15s}: ${coef:>10,.0f} per unit")
```

```python
# Predictions
y_pred = model_multi.predict(X)

# Metrics
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y, y_pred)

print(f"\nModel Performance:")
print(f"  RMSE: ${rmse:,.0f}")
print(f"  R² Score: {r2:.4f}")

# Example prediction
new_house = pd.DataFrame({
    'sqft': [2500],
    'bedrooms': [4],
    'age': [5]
})

prediction = model_multi.predict(new_house)[0]
print(f"\n🏠 Prediction: A 2500 sqft, 4-bedroom, 5-year-old house costs \${prediction:,.0f}")
```

## Train-Test Split 🔀

```python
from sklearn.model_selection import train_test_split

# Generate larger dataset
X_large = df[['sqft', 'bedrooms', 'age']]
y_large = df['price']

# Split: 70% train, 30% test
X_train, X_test, y_train, y_test = train_test_split(
    X_large, y_large, test_size=0.3, random_state=42
)

print(f"Training set: {len(X_train)} samples")
print(f"Test set: {len(X_test)} samples")

# Train on training set
model_split = LinearRegression()
model_split.fit(X_train, y_train)

# Evaluate on both
train_r2 = model_split.score(X_train, y_train)
test_r2 = model_split.score(X_test, y_test)

print(f"\nTraining R²: {train_r2:.4f}")
print(f"Test R²:     {test_r2:.4f}")
print(f"\n✅ If close: Model generalizes well")
print(f"❌ If test much lower: Model overfitted to training data")
```

## Visualizing Predictions 📉

```python
# Actual vs Predicted
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Actual vs Predicted
predictions = model_multi.predict(X)
axes[0].scatter(y, predictions, alpha=0.7, s=100, color='blue')
axes[0].plot([y.min(), y.max()], [y.min(), y.max()], 'r--', lw=2, label='Perfect predictions')
axes[0].set_xlabel('Actual Price')
axes[0].set_ylabel('Predicted Price')
axes[0].set_title('Actual vs Predicted Prices')
axes[0].legend()
axes[0].grid(True, alpha=0.3)
axes[0].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x/1e6:.1f}M' if x >= 1e6 else f'${x/1e3:.0f}K'))
axes[0].yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x/1e6:.1f}M' if x >= 1e6 else f'${x/1e3:.0f}K'))

# Plot 2: Residuals
residuals = y.values - predictions
axes[1].scatter(predictions, residuals, alpha=0.7, s=100, color='green')
axes[1].axhline(y=0, color='r', linestyle='--', lw=2)
axes[1].set_xlabel('Predicted Price')
axes[1].set_ylabel('Residual')
axes[1].set_title('Residual Plot')
axes[1].grid(True, alpha=0.3)
axes[1].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x/1e6:.1f}M' if x >= 1e6 else f'${x/1e3:.0f}K'))
axes[1].yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x/1e3:.0f}K'))

plt.tight_layout()
plt.show()

print("Left plot: Points on diagonal = good predictions")
print("Right plot: Points near zero line = small errors")
```

## Key Concepts 📚

### Simple: y = mx + b
- m (slope): Rate of change
- b (intercept): Starting value

### Multiple: y = b₀ + b₁x₁ + b₂x₂ + ... + bₙxₙ
- Each feature gets its own coefficient
- More flexible but harder to visualize

### Metrics
- **R²:** Explained variance (0-1, higher better)
- **RMSE:** Prediction error magnitude
- **MAE:** Mean absolute error

## Key Takeaways 🎯

✅ Linear regression fits a line through data

✅ Used for continuous predictions

✅ Easy to interpret (coefficients show impact)

✅ Check R², RMSE to evaluate fit

✅ Works best for linear relationships
